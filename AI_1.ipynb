—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.


```python
from matplotlib import pyplot
import numpy


values = numpy.random.sample(15)
indices = list(range(len(values)))

sr = numpy.mean(values)
md = numpy.median(values)

pyplot.figure(figsize=(10, 5))
pyplot.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
pyplot.ylabel('–ò–Ω–¥–µ–∫—Å —ç–ª–µ–º–µ–Ω—Ç–∞')
pyplot.scatter(values, indices)
pyplot.show()


print('–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: ', sr, '\n–°—Ä–µ–¥–Ω–µ–µ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: ', md)

if sr > md:
    print('–°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –±–æ–ª—å—à–µ')
elif sr < md:
    print('–°—Ä–µ–¥–Ω–µ–µ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –±–æ–ª—å—à–µ')
else:
    print('–ó–Ω–∞—á–µ–Ω–∏—è —Ä–∞–≤–Ω—ã')
```

**–ó–∞–¥–∞–Ω–∏–µ 4.3.2.** –î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è $\frac{\sqrt{1+e^{\sqrt{x}} +\cos x^{2}}}{|1-\sin^{3} |} +\ \ln |2x|$; –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏
(ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏
–¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì –ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π}.

```python
from matplotlib import pyplot
import numpy


y = []

for x in range(1, 11):
    y.append(numpy.sqrt(1 + numpy.e ** numpy.sqrt(x) + numpy.cos(x ** 2)) / abs(1 - numpy.sin(x) ** 3) + numpy.log(abs(2 * x)))


pyplot.plot(y)
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.show()

pyplot.scatter([1, 2, 3, 4, 5],  y[:len(y) // 2])
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.show()
```

**–ó–∞–¥–∞–Ω–∏–µ 4.3.5.** –°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è
–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: $ùëí^{(ùë•+ùë¶)}, sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•^ùë¶$

```python
from math import *


def func(x, y, do):
  if do == 1:
    return x + y
  if do == 2:
    return x - y
  if do == 3:
    return x * y
  if do == 4:
      if y == 0:
          print('–ù–∞ 0 –¥–µ–ª–∏—Ç—å –Ω–µ–ª—å–∑—è')
          return
      return x / y
  if do == 5:
    return e ** (x + y)
  if do == 6:
    return sin(x + y)
  if do == 7:
    return cos(x + y)
  if do == 8:
    return x ** y


while (True):
    x = float(input('–í–≤–µ–¥–∏—Ç–µ x: '))
    y = float(input('–í–≤–µ–¥–∏—Ç–µ y: '))

    print('\n1. +\n2. -\n3. *\n4. /\n5. e^x+y\n6. sin(x+y)\n7. cos(x+y)\n8. x^y')
    do = int(input('–í—ã–±–µ—Ä–∏—Ç–µ –æ–ø–µ—Ä–∞—Ü–∏—é: '))
    print('–û—Ç–≤–µ—Ç:', func(x, y, do))
```
